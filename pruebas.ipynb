2. Limpieza y transformación basica

```python
df.isna().sum()
```

```python
##Elimino la columna id porque no aporta nada a mi análisis y tiene muchos valores nulos.
df.drop('id', axis=1, inplace=True, errors='ignore')
```

```python
##Eliminamos todas las filas que contienen valores nulos, ya que no son muchas y no queremos imputar datos.
df.dropna(inplace=True)
```

```python
pulgadas = df['pulgadas'].unique()
print(pulgadas)
```

```python
###Convertimos la columna 'pulgadas' a tipo numérico, forzando los errores a NaN para manejar cualquier valor no convertible.
df['pulgadas'] = pd.to_numeric(df['pulgadas'], errors='coerce')
```

```python
##Corregimos el error de tipeo en la columna tipo_laptop, cambiando Netbook por Notebook
df['tipo_laptop'] = df['tipo_laptop'].str.replace('Netbook', 'Notebook', regex=False)
```

```python
df['compania'].value_counts()
```

```python
##Cambiamos la referncia de Vero por Acer, ya que es un error de tipeo y no existe la marca Vero y es una linea de la marca Acer
df['compania'] = df['compania'].str.replace('Vero', 'Acer', regex=False)
```

```python
##Cambiamos peso a tipo numérico, eliminando la unidad 'kg' y convirtiendo a float
df['peso'] = df['peso'].str.replace('kg', '', regex=False)

##Cambiamos el precio a tipo numérico, eliminando el símbolo de moneda y convirtiendo a float
```

```python
##Revisamos los valores únicos de la columna peso para verificar que se hayan limpiado correctamente y que no haya quedado ningún valor no numérico.
peso = df['peso'].unique()
print(peso)


 ##'2.54' '3.42' '5.8' '1.28' '2.33' '1.45' '2.79' '8.23' '1.26' '1.84'
 ##'0.0002' '2.6' '2.26' '3.25' '1.59' '1.13' '1.42' '1.78' '1.10' '1.15'
 ##'1.27' '1.43' '2.31' '1.16' '1.64' '2.17' '1.47' '3.78' '1.79' '0.91'
 
```

```python
##Como hay un valor de 0.0002 kg, que es claramente un error de tipeo, lo reemplazamos por 0.2 kg, que es un valor más razonable para el peso de una laptop.
df['peso'] = df['peso'].str.replace('0.0002', '2.0', regex=False)
```

```python
peso= df['peso'].unique()
print(peso)
```

```python
##Cambiamos el precio a tipo numérico, eliminando el símbolo de moneda y convirtiendo a float
##df['precio_Kg'] = pd.to_numeric(df['precio_Kg'].astype(str).str.replace('₹', '', regex=False).str.replace(',', '', regex=False),errors='coerce')
df['precio'] = pd.to_numeric(df['precio'].astype(str).str.replace('₹', '', regex=False).str.replace(',', '', regex=False), errors='coerce')
```

Hacemos separacion por sistema operativo

```python
sistema_operativo = df['sistema_operativo'].unique()
print(sistema_operativo)

##Lista de sistemas operativos que habian
##'macOS' 'No OS' 'Windows 10' 'Mac OS X' 'Linux' 'Windows 10 S'
##'Chrome OS' 'Windows 7' 'Android'

```

```python
##Unificamos los sistemas operativos Windows 10 y Windows 11 en una sola categoría llamada Windows, ya que ambos son versiones del mismo sistema operativo y no queremos que la variable sistema_operativo tenga demasiadas categorías diferentes que puedan dificultar el análisis.
df['sistema_operativo'] = df['sistema_operativo'].str.replace('Windows 10', 'Windows', regex=False)
df['sistema_operativo'] = df['sistema_operativo'].str.replace('Windows 7', 'Windows', regex=False)
df['sistema_operativo'] = df['sistema_operativo'].str.replace('Windows S', 'Windows', regex=False)
df['sistema_operativo'] = df['sistema_operativo'].str.replace('Windows 11', 'Windows', regex=False)     
df['sistema_operativo'] = df['sistema_operativo'].str.replace('Mac OS X', 'macOS', regex=False)
```
