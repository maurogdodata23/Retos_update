2. Limpieza y transformacioiin basica

```python
df.isna().sum()
```

```python
##Elimino la columna id porque no aporta nada a mi análisis y tiene muchos valores nulos.
df.drop('id', axis=1, inplace=True, errors='ignore')
```

```python
##Eliminamos todas las filas que contienen valores nulos, ya que no son muchas y no queremos imputar datos.
df.dropna(inplace=True)
```

```python
###Dell Precision 7520 FHD 15.6" Workstation Business PC (Intel Quad Core i7-7700HQ, 16GB Ram, 256GB SSD, Thunderbolt 3) Nvidia Quadro M1200 4GB DDR5
df['pulgadas'] = df['pulgadas'].str.replace('?', '15.6', regex=False)
```

```python
##Corregimos el error de tipeo en la columna tipo_laptop, cambiando Netbook por Notebook
df['tipo_laptop'] = df['tipo_laptop'].str.replace('Netbook', 'Notebook', regex=False)
```

```python
df['compania'].value_counts()
```

```python
##Cambiamos la referncia de Vero por Acer, ya que es un error de tipeo y no existe la marca Vero y es una linea de la marca Acer
df['compania'] = df['compania'].str.replace('Vero', 'Acer', regex=False)
```

```python
##Miramos si hay filas duplicadas, ya que pueden sesgar nuestro análisis y no aportan nada nuevo.
df.duplicated().sum()
```

```python
#Eliminamos las filas duplicadas, ya que no aportan nada nuevo a nuestro análisis y pueden sesgar los resultados.
df.drop_duplicates(inplace=True)
```
